# Workers With The Highest Salaries

Find the titles of workers that earn the highest salary. Output the highest-paid title or multiple titles that share the highest salary.

```python
title = pd.read_csv('title.csv')
worker = pd.read_csv('worker.csv')
```

```python
# merge 2 tables
pd.merge(worker,title,left_on='worker_id',right_on='worker_ref_id',how='left')
```

```python
# create dataframe
df = pd.merge(worker,title,left_on='worker_id',right_on='worker_ref_id',how='left')
```

```python
# find the max salary of each worker tiles
(pd.merge(worker,title,left_on='worker_id',right_on='worker_ref_id',how='left')).groupby(['worker_title'])['salary'].max()
```

```python
# set a dataframe from max salary per worker titles
((pd.merge(worker,title,left_on='worker_id',right_on='worker_ref_id',how='left')).groupby(['worker_title'])['salary'].max()).reset_index()
```

find out the worker who has the max salary of each tiles

```python
# create the max salary per title column
df.groupby('worker_title').transform(max)['salary']
```

```python
# add new column inton dataframe next to 'salary' column
df.insert(loc=4,column='Max salary of each title',value=df.groupby('worker_title').transform(max)['salary'])
```

```python
# filter the workers whose salary is equal to max salary of each title and drop 'Max salary of each title'
df[df['salary']==df['Max salary of each title']].drop(columns={'Max salary of each title'})
```
