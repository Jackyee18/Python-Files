```python
import pandas as pd
import numpy as np
q = pd.read_csv('doordash_delivery.csv')
```

# Avg Order Cost During Rush Hours
Write a query that returns the average order cost per hour during hours 3 PM -6 PM (15-18) in San Jose. For calculating time period use 'Customer placed order datetime' field. Earnings value is 'Order total' field. Order output by hour.

```python
q
```

```python
# filter table with delivery_region in 'San Jose'
q[q['delivery_region']=='San Jose']
```

```python
# change from object into datetime and get the time of the 'customer_placed_order_datetime' column
q['customer_placed_order_datetime'].apply(pd.to_datetime).dt.hour
q['placed_order_hour'] = q['customer_placed_order_datetime'].apply(pd.to_datetime).dt.hour
```

```python
# condition of order time is between 15 and 18
(q['placed_order_hour']>=15)&(q['placed_order_hour']<=18)
```

```python
q[(q['delivery_region']=='San Jose')&(q['placed_order_hour']>=15)&(q['placed_order_hour']<=18)]
```

```python
# the total order of this tables
q[(q['delivery_region']=='San Jose')&(q['placed_order_hour']>=15)&(q['placed_order_hour']<=18)]['order_total'].sum()
```

```python
# the average amount of Rush hour (15-18)
q[(q['delivery_region']=='San Jose')&(q['placed_order_hour']>=15)&(q['placed_order_hour']<=18)]['order_total'].sum()/(18-15)
```
