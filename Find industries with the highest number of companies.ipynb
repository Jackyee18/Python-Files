## Find the most popular sector in the Forbes list

```python
df = pd.read_csv('forbes_global_2010_2014.csv')
```

```python
# solution with groupby function
df.groupby('sector')['sector'].count()
```

```python
# solution with value_counts syntax
df.sector.value_counts()
```

## Find the most profitable company in the financial sector of the entire world along with its continent

```python
# filter the dataframe with 'Financials' in Sector
df[df.sector=='Financials']
```

```python
# sort values by profit in descending
df[df.sector=='Financials'].sort_values(by='profits',ascending=False)
```

```python
# top 5 most profitable company in the financial sector of the entire world
df[df.sector=='Financials'].sort_values(by='profits',ascending=False).iloc[0:6,[0,3,4,6,7,8]]
```

## Find the most profitable company in the financial sector of each continent

```python
# create a table filter with Financials sector and sort by profit in descending
df1= df[df.sector=='Financials'].sort_values(by='profits',ascending=False)
```

```python
# create a max profit of each continent
df1.profits==(df1.groupby('continent').transform(max))['profits']
```

```python
# final solution
df1[df1.profits==(df1.groupby('continent').transform(max))['profits']]
```

## Find industries with the highest number of companies. Output the industry along with the number of companies. Sort records based on the number of companies in descending order.

Output the sector along with the number of companies.

```python
df.industry.value_counts()
```

```python
# rename 2 two columns
df.industry.value_counts().reset_index().rename(columns={'index':'industry','industry':'number of companies'})
```

## Find industries that make a profit

Find all industries with a positive average profit. For those industries extract their lowest sale.
Output the industry along with the corresponding lowest sale and average profit.
Sort the output based on the lowest sales in ascending order.

```python
(df.groupby(['industry']).agg({'profits':np.mean,'sales':np.min})).rename(columns={'profits':'avg profit','sales':'min sale of industry'})
```
