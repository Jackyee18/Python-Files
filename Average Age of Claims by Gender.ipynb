```python
import pandas as pd
import numpy as np
claims=pd.read_csv('cvs_claims.csv')
accounts=pd.read_csv('cvs_accounts.csv')
```

```python
q=pd.merge(claims,accounts)
q
```

You have been asked to calculate the average age by gender of people who filed more than 1 claim in 2021.


The output should include the gender and average age rounded to the nearest whole number.

```python
# change data type and add the year column
(q.date_submitted.apply(pd.to_datetime)).dt.year
q['year']=(q.date_submitted.apply(pd.to_datetime)).dt.year
```

```python
# filter with data in 2021
q=q[q.year==2021]
```

```python
# find the total claim by gender
q.groupby(['account_id','gender'])['claim_id'].count().reset_index().rename(columns={'claim_id':'total_claims'})
q=q.groupby(['account_id','gender'])['claim_id'].count().reset_index().rename(columns={'claim_id':'total_claims'})
```

```python
# find accounts have more than 1 claim and merge to find the age of them
q[q.total_claims>1].merge(accounts).drop(columns=['total_claims'])
q=q[q.total_claims>1].merge(accounts).drop(columns=['total_claims'])
```

```python
# the average age by gender of people who filed more than 1 claim in 2021
q.groupby(['gender'])['age'].mean()
```
