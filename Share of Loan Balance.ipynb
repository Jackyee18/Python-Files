```python
import pandas as pd
import numpy as np
q=pd.read_csv('submissions.csv')
q
```

Write a query that returns the rate_type, loan_id, loan balance , and a column that shows with what percentage the loan's balance contributes to the total balance among the loans of the same rate type.

```python
q1=q.groupby('rate_type')['balance'].sum().reset_index()
q1
```

```python
q3=q.merge(q1,left_on='rate_type',right_on='rate_type',suffixes=('', '_per_rate_type'))
```

```python
q3['pct_as_rate_type']=(q3['balance']/q3['balance_per_rate_type']*100).round(2)
```
