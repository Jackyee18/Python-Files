## Find the total cost of each customer's orders. Output customer's id, first name, and the total order cost. Order records by customer's first name alphabetically.

```python
# find the total cost of each customer's order (using order table)
orders.groupby('cust_id')['total_order_cost'].sum()
```

```python
# create a total order cost by customer id
total_order_cost_by_customer_id = orders.groupby('cust_id')['total_order_cost'].sum().reset_index()
```

```python
# join with customer table to add name column
total_order_cost_by_customer_id.merge(customers,how='left',left_on='cust_id',right_on='id')
```

```python
# drop some columns as requirements
result = total_order_cost_by_customer_id.merge(customers,how='left',left_on='cust_id',right_on='id')
result.drop(columns=['last_name','city','address','phone_number'])
```

```python
# final solutions:
(((orders.groupby('cust_id')['total_order_cost'].sum()).reset_index()).merge(customers,how='left',left_on='cust_id',right_on='id')).drop(columns=['last_name','city','address','phone_number'])
```
