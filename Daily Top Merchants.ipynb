```python
import pandas as pd
merchants = pd.read_csv('doordash_merchants.csv')
orders = pd.read_csv('doordash_orders.csv')
```

## Daily Top Merchants

For each day, find the top 3 merchants with the highest number of orders on that day. In case of a tie, multiple merchants can share the same place but on each day, there should always be at least 1 merchant on the first, second and third place.
Output the date, the name of the merchant and their place in the daily ranking.

```python
# check the datatype of order_timestamp
orders.order_timestamp.dtype
```

```python
# change from object into datetime
orders.order_timestamp.apply(pd.to_datetime)
```

```python
# get the date from datetime
(orders.order_timestamp.apply(pd.to_datetime)).dt.date
```

```python
# overwrite the order date with the existing column
orders.order_timestamp=(orders.order_timestamp.apply(pd.to_datetime)).dt.date
```

```python
# group by with 2 values order_timestamp, merchant
orders.groupby(['order_timestamp','merchant_id']).count()['customer_id']
```

```python
# set dataframe
df = orders.groupby(['order_timestamp','merchant_id']).count()['customer_id'].reset_index()
orders.groupby(['order_timestamp','merchant_id']).count()['customer_id'].reset_index()
```

```python
# rename the columns to better understand and sort value by order_date and total customers
(df.rename(columns={'order_timestamp':'order_date','customer_id':'total customers'})).sort_values(by=['order_date','total customers'],ascending=False)
```

```python
# set a new dataframe
df1 = (df.rename(columns={'order_timestamp':'order_date','customer_id':'total customers'})).sort_values(by=['order_date','total customers'],ascending=False)
```

```python
# find the rank of 'total customers' in descending group by 'order date' 
df1.groupby(['order_date'])['total customers'].rank(method='first',ascending=True)
```

```python
# set into dataframe to check
df1['rank'] = df1.groupby(['order_date'])['total customers'].rank(method='first',ascending=False)
df1
```

```python
# find the top 3 merchants which have the most total customers of each order date
df1[df1['rank']<=3]
```

```python
# drop the redundant columns
df1[df1['rank']<=3].drop(columns={'rank'})
```

```python
# set a new dataframe
df2 = df1[df1['rank']<=3].drop(columns={'rank'})
df1[df1['rank']<=3]
```

```python
# merge 2 table and drop 'id' and 'zipcode' columns
pd.merge(df2,merchants,left_on='merchant_id',right_on='id',how='left').drop(columns={'id','zipcode'})
```
