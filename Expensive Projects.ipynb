Given a list of projects and employees mapped to each project, calculate by the amount of project budget allocated to each employee . The output should include the project title and the project budget rounded to the closest integer. Order your list by projects with the highest budget per employee first.

```python
project_df = pd.read_csv('ms_projects.csv')
emp_pro_df = pd.read_csv('ms_emp_projects.csv')
```

```python
# count the employee of each project
emp_pro_df.groupby('project_id')['emp_id'].count()
```

```python
# set an index to create a dataframe
emp_pro_df.groupby('project_id')['emp_id'].count().reset_index()
```

```python
# rename the 'emp_id' into 'emp_count' of dataframe
(emp_pro_df.groupby('project_id')['emp_id'].count().reset_index()).rename(columns={'emp_id':'emp_count'})
```

```python
# merge with the project table to add project title
(emp_pro_df.groupby('project_id')['emp_id'].count().reset_index()).rename(columns={'emp_id':'emp_count'}).merge(project_df,how='left',left_on='project_id',right_on='id')
```

```python
# set a dataframe of result
df = (emp_pro_df.groupby('project_id')['emp_id'].count().reset_index()).rename(columns={'emp_id':'emp_count'}).merge(project_df,how='left',left_on='project_id',right_on='id')
```

```python
# drop a redundant column
df = df.drop(columns=['id'])
```

```python
# check the type of columns before making an aggregation function
df.info()
```

```python
# create the 'budget per employee' column
df['budget per employee'] = df['budget']/df['emp_count']
```

```python
# round the number to the closest int
df['budget per employee'].round()
```

```python
# change from 'float' into 'int'
(df['budget per employee'].round()).astype(int)
```

```python
# set the new values into the existing column
df['budget per employee'] = (df['budget per employee'].round()).astype(int)
```

```python
# sort by 'budget per employee' 
df.sort_values(by='budget per employee',ascending=False)
```
