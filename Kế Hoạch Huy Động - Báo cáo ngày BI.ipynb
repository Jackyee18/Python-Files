## CƠ CẤU TỔNG HUY ĐỘNG TRONG FILE KẾ HOẠCH
TỔNG HUY ĐỘNG (CF) = HUY ĐỘNG KKH VNĐ(FL) + HUY ĐỘNG KKH NT KHÁC(HB) +  HUY ĐỘNG CKH VNĐ (KH) +  HUY ĐỘNG CKH NT (LX) +  HĐ tay CKH (NN) +  HĐ tay KKH (PD) +  GTCG (QT)


```python
import pandas as pd

def process_filtered_df(filtered_df, column_pattern):
    # Generate the new column names
    new_column_names = [column_pattern.format(i) for i in range(1, 13)] # đặt tên cột theo tên chỉ tiêu kèm số tháng phía sau
    # Rename the columns
    filtered_df = filtered_df.rename(columns=dict(zip(filtered_df.columns, new_column_names)))

    a = df.iloc[4:432, 1].reset_index(drop=True) # cột mã đơn vị
    b = filtered_df.reset_index(drop=True) # 12 cột của 1 chỉ tiêu kế hoạch 12 tháng

    result = pd.concat([a, b], axis=1) # ghép cột chỉ tiêu kế hoạch 12 tháng với cột mã đơn vị 
    result = result.rename(columns={1: 'BRANCH_NO'})
    melted_df = result.melt(id_vars='BRANCH_NO', var_name='Variable', value_name='TARGET') # chuyển format 13 cột => 3 cột
    # Split the column
    melted_df[['Category', 'MONTH']] = melted_df['Variable'].str.split(' _ ', expand=True) # tách 2 cột
    melted_df=melted_df.drop(columns=['Variable','Category'])
    melted_df=melted_df.fillna(0) # thêm giá trị 0 vào các ô trống
    
    return melted_df
```

```python
# cập nhập kế hoạch của khối CMB
df=pd.read_excel(file_path,sheet_name='CN (TĐ NIM)')
# Define the repetitive information as a dictionary
data_info = {
    1: {'columns': (164, 176), 'pattern': "HUY ĐỘNG KKH VNĐ - CN kế hoạch _ {}", 'term': 'KKH', 'currency': 'VND'},
    2: {'columns': (206, 218), 'pattern': "HUY ĐỘNG KKH NT KHÁC - CN kế hoạch _ {}", 'term': 'KKH', 'currency': 'KHAC'},
    3: {'columns': (290, 302), 'pattern': "HUY ĐỘNG CKH VNĐ - CN kế hoạch _ {}", 'term': 'CKH', 'currency': 'VND'},
    4: {'columns': (332, 344), 'pattern': "HUY ĐỘNG CKH NT  - CN kế hoạch _ {}", 'term': 'CKH', 'currency': 'KHAC'},
    }


# Process each part and store them in a list
data_parts = []
for i, info in data_info.items():
    filtered_df = df.iloc[4:432, info['columns'][0]:info['columns'][1]]
    column_pattern = info['pattern']
    part = process_filtered_df(filtered_df, column_pattern)
    part['TERM'] = info['term']
    part['CURRENCY'] = info['currency']
    data_parts.append(part)

# Concatenate the parts and filter the resulting dataframe
data = pd.concat(data_parts)
KH_CN_df = data[data['BRANCH_NO'] != 0].fillna(0)
KH_CN_df['BUSINESS_LINE']='CN'
KH_CN_df
```
