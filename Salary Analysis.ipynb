#  Employees Without Benefits
Find the ratio between the number of employees without benefits to total employees. Output the job title, number of employees without benefits, total employees relevant to that job title, and the corresponding ratio. Order records based on the ratio in ascending order.


```python
q1=q[q.benefits==0].groupby('jobtitle')['id'].count().reset_index()
q2=q[q.benefits!=0].groupby('jobtitle')['id'].count().reset_index()
q3=pd.merge(q1,q2,on='jobtitle',how='left').rename(columns={'id_x':'without_benefits','id_y':'with_benefits'})
q3['ratio']=(100*q3['without_benefits']/q3['with_benefits']).round(2)
q3
```

# Highest And Lowest Paying Jobs


Find the ratio and the difference between the highest and lowest total pay for each job title. Output the job title along with the corresponding difference, ratio, highest total pay, and the lowest total pay. Sort records based on the ratio in descending order.



```python
q1=(q.groupby('jobtitle').totalpay.max() / q.groupby('jobtitle').totalpay.min()).reset_index().merge((q.groupby('jobtitle').totalpay.max() - q.groupby('jobtitle').totalpay.min()),on='jobtitle',suffixes=('_ratio', '_diff'))
```

```python
# Replace inf with Zero in dataframe
q1.replace([np.inf,-np.inf],0,inplace=True)
```

```python
q_max=q.groupby('jobtitle').totalpay.max()
q_min=q.groupby('jobtitle').totalpay.min()
pd.merge(q_max,q_min,on='jobtitle',suffixes=('_max', '_min')).merge(q1,on='jobtitle').sort_values(by='totalpay_ratio',ascending=False)
```

Find all people who earned more than the average in 2013 for their designation but were not amongst the top 5 earners for their job title. Use the totalpay column to calculate total earned and output the employee name(s) as the result.

# Make a pivot table to find the highest payment in each year for each employee

Make a pivot table to find the highest payment in each year for each employee.
Find payment details for 2011, 2012, 2013, and 2014.
Output payment details along with the corresponding employee name.
Order records by the employee name in ascending order

```python
# find the highest payment in each year for each employee
q['max_totalpaybenefits']=q.groupby('year')['totalpaybenefits'].transform('max')
# sort out with this codition to find the result
q.loc[q['totalpaybenefits']==q['max_totalpaybenefits'],['id','employeename','year']].sort_values(by='employeename',ascending=True)
```

```python
q.pivot_table(index='employeename',columns='year',values='totalpaybenefits').sort_values(by='employeename',ascending=True)
```
