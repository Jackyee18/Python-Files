```python
import pandas as pd
import numpy as np
orders=pd.read_csv('doordash_orders.csv')
merchants=pd.read_csv('doordash_merchants.csv')
```

For each merchant, find how many orders and first-time orders they had. First-time orders are meant from the perspective of a customer and are the first order that a customer ever made. In order words, for how many customers was this the first-ever merchant they ordered with?
Output the name of the merchant, the total number of their orders and the number of these orders that were first-time orders.

```python
# find the total customers of each merchant
orders.groupby('merchant_id')['customer_id'].count().reset_index().rename(columns={'customer_id':'total_customers'})
q1=orders.groupby('merchant_id')['customer_id'].count().reset_index().rename(columns={'customer_id':'total_customers'})
q1
```

```python
# first time customer condition is equal with values being duplicated with 'customer_id' and 'merchant_id':
orders.duplicated(subset=['customer_id','merchant_id'])
```

```python
# orders of customers first time order with merchant is equivalent is the opposite with the aboved codition: 
orders[~orders.duplicated(subset=['customer_id','merchant_id'])]
```

```python
# find the total first time customers of each merchant
orders[~orders.duplicated(['customer_id','merchant_id'])].groupby('merchant_id')['customer_id'].count().reset_index().rename(columns={'customer_id':'total_first_time_customers'})
q2=orders[~orders.duplicated(['customer_id','merchant_id'])].groupby('merchant_id')['customer_id'].count().reset_index().rename(columns={'customer_id':'total_first_time_customers'})
q2
```

```python
q=pd.merge(q1,q2)
q
```

```python
pd.merge(q,merchants,left_on='merchant_id',right_on='id').drop(columns=['id','zipcode'])
```
