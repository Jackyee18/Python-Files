```python
participants_df = pd.read_csv('google_competition_participants.csv')
scores_df = pd.read_csv('google_competition_scores.csv')
```

Find the winning teams of DeepMind employment competition.
Output the team along with the average team score.
Sort records by the team score in descending order.

```python
# merge 2 tables
pd.merge(participants_df,scores_df)
```

```python
# create a dataframe
df = pd.merge(participants_df,scores_df)
```

```python
# the average score by each team 
df.groupby('team_id')['member_score'].mean()
```

```python
# reset index to set a new dataframe
(df.groupby('team_id')['member_score'].mean()).reset_index()
```

```python
#  rename the column to better understand
((df.groupby('team_id')['member_score'].mean()).reset_index()).rename(columns={'member_score':'avg_team_score'})
```

```python
#  sort value by descending 
(((df.groupby('team_id')['member_score'].mean()).reset_index()).rename(columns={'member_score':'avg_team_score'})).sort_values(by='avg_team_score',ascending=False)
```

```python
# Final solution:
((df.groupby('team_id')['member_score'].mean()).reset_index().sort_values(by='member_score',ascending=False)).rename(columns={'member_score':'avg_team_score'})
```
